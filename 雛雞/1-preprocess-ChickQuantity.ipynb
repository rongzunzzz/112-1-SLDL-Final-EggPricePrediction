```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
df = pd.read_excel('../data_original/GetResource.xls', sheet_name='107年週統計', header=1)
```

```python
df.head(7)
```

```python
# 重新命名欄位名稱
new_column_names = {'數量\類別': '年', 'Unnamed: 1': '月', 'Unnamed: 2': '週'}
df_new = df.rename(columns=new_column_names)

# 選擇要保留的欄位
columns_to_keep = ['年', '月', '週', '白殼蛋雛雞', '紅殼蛋雛雞', '進口母雛數']
df_new = df_new[columns_to_keep]

# 移除不必要的欄位
df_new.drop(0, inplace=True)
df_new.drop(df.index[-2:], inplace=True)

df_new
```

```python
df_new['年'] = df_new['年'].fillna(method='ffill')
df_new['月'] = df_new['月'].fillna(method='ffill')
```

```python
df_new
```

```python
df_test = df_new
```

```python
import pandas as pd
from datetime import datetime, timedelta

# 生成 "Date" 欄位
df_test['Date'] = df_test.apply(lambda row: datetime(1911 + row['年'], row['月'], 1) + timedelta(weeks=row['週'] - 1), axis=1)

# 將 "Date" 欄位設置為索引
df_test.set_index('Date', inplace=True)

# 重新取樣以填充所有相鄰日期之間的缺失值
df_resampled = df_test.resample('D').ffill()

# 如果是每個月 1 號，則將進口母雞數填入整個月份
df_resampled['進口母雛數'] = df_resampled.groupby(df_resampled.index.to_period('M'))['進口母雛數'].transform('first')

# 重設索引
df_resampled.reset_index(inplace=True)


# 顯示結果
display(df_resampled)
```

```python
# 移除 '年'、'月'、'週' 這三個欄位
df_resampled = df_resampled.drop(['年', '月', '週'], axis=1)

new_column_names = {'白殼蛋雛雞': 'WhiteChickQuantity', '紅殼蛋雛雞': 'RedChickQuantity', '進口母雛數': 'ImportChickenQuantity'}
df_output = df_resampled.rename(columns=new_column_names)

display(df_output)
```

```python
# 要增加的日期範圍
# new_dates = [datetime(2019, 12, 23), datetime(2019, 12, 24), datetime(2019, 12, 25), datetime(2019, 12, 26), datetime(2019, 12, 27), datetime(2019, 12, 28), datetime(2019, 12, 29), datetime(2019, 12, 30), datetime(2019, 12, 31)]
new_dates = [datetime(2018, 12, 30), datetime(2018, 12, 31)]

# 使用 for 迴圈生成每一行並添加到 DataFrame
for date in new_dates:
    new_row = {'Date': date,
               'WhiteChickQuantity': df_output['WhiteChickQuantity'].iloc[-1],  # 取上一行的數據
               'RedChickQuantity': df_output['RedChickQuantity'].iloc[-1],        # 取上一行的數據
               'ImportChickenQuantity': df_output['ImportChickenQuantity'].iloc[-1]  # 取上一行的數據
               }
    df_output = df_output.append(new_row, ignore_index=True)

# 重設索引
df_output.reset_index(drop=True, inplace=True)

display(df_output)
```

```python
df_output.to_csv('2018ChickQuantity.csv', index=False)
```
