```python
import pandas as pd
import numpy as np
import copy
from datetime import datetime, timedelta
```

```python
df = pd.read_excel('../data_original/GetResource-2.xls')
```

```python
df
```

```python
df_year = df[df['月份'].notna() & df['月份'].str.startswith('107')].reset_index(drop=True)
display(df_year)
```

```python
# 刪除最後兩個不需要的columns
df_year.drop(df_year.columns[-2:], axis=1, inplace=True)
df_year
```

```python
df_test = copy.deepcopy(df_year)
```

```python
# 先將 '年'、'月'、'日' 中文字符替換為空白
df_test['月份'] = df_test['月份'].str.replace('107年', '2018').str.replace('月', '')

# 將 '月份' 轉換成西元年
# df_test['月份'] = df_test['月份'].astype(int) + 191100

# 將 '月份' 轉換為日期範圍
df_test['月份'] = pd.to_datetime(df_test['月份'].astype(str) + '01', format='%Y%m%d')

# 顯示結果
display(df_test)
```

```python
# df_test.iloc[0, 0] = datetime(2022, 1, 1)
# display(df_test)
```

```python
new_row = copy.deepcopy(df_test.iloc[-1])
new_row['月份'] = datetime(2018, 12, 31)
df_test = df_test.append(new_row, ignore_index=True)
df_test
```

```python
# 將 "月份" 欄位設置為索引
df_test.set_index('月份', inplace=True)

# 重新取樣以填充所有相鄰日期之間的缺失值
df_resampled = df_test.resample('D').ffill()

# 重設索引
df_resampled.reset_index(inplace=True)
```

```python
df_resampled
```

```python
df_resampled.to_csv('./Chicken/2018Chicken.csv', index=False)
```
